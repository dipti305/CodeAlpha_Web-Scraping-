```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load Titanic dataset
df = sns.load_dataset("titanic")
print(df.head())
```

```python
# Clean slice and copy
df_clean = df.dropna(subset=['age', 'embarked']).copy()

# Cast to object (string) type before mapping
df_clean['survived'] = df_clean['survived'].astype('object').map({0: 'No', 1: 'Yes'})
```

```python
sns.countplot(x='sex', hue='survived', data=df_clean)
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
print("Insight: Most passengers were aged 20–40, but children were also onboard")
```

```python
sns.histplot(df_clean['age'], kde=True, bins=30)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
print("Insight: Most passengers were aged 20–40, but children were also onboard")
```

```python
sns.boxplot(x='pclass', y='age', data=df_clean)
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
print(" Insight: 1st class passengers were generally older than 3rd class")
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_clean.select_dtypes(include=np.number).corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Between Numerical Features")
plt.show()
print(" Insight: Slight negative correlation between age and survival")
```

```python
df_clean['embarked'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Embarkation Port Distribution")
plt.ylabel("")
plt.show()
print(" Insight: Most passengers boarded from Southampton (S)")
```
